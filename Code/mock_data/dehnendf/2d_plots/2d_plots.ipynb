```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = '../main_program_results/dehnendf_psp.npz'
data = np.load(filename)
centres = data['cluster']
max_dots = np.nanmax(np.abs(data['result']), axis=1)

# filter out nan
centres = centres[~np.isnan(max_dots)]
max_dots = max_dots[~np.isnan(max_dots)]
```

```python
plt.figure(figsize=(10,8), facecolor='black')
plt.style.use("dark_background")
plt.scatter(*centres.T[[0,1]], c=max_dots, marker='.', s=5, cmap='plasma', vmin=0, vmax=1)
plt.colorbar(label='Maximum Absolute Dot Product')
plt.xlabel('$x/R_0$')
plt.ylabel('$y/R_0$')
plt.title('Dehnen DF with PowerSphericalPotential, 1.7 Million Stars')
plt.show()
```

```python
plt.figure(figsize=(10,8), facecolor='black')
plt.style.use("dark_background")
plt.scatter(*centres.T[[0,2]], c=max_dots, marker='.', s=5, cmap='plasma', vmin=0, vmax=1)
plt.colorbar(label='Maximum Absolute Dot Product')
plt.xlabel('$x/R_0$')
plt.ylabel('$z/R_0$')
plt.title('Dehnen DF with PowerSphericalPotential, 1.7 Million Stars')
plt.show()
```

```python
plt.figure(figsize=(10,8), facecolor='black')
plt.style.use("dark_background")
plt.scatter(*centres.T[[1,2]], c=max_dots, marker='.', s=5, cmap='plasma', vmin=0, vmax=1)
plt.colorbar(label='Maximum Absolute Dot Product')
plt.xlabel('$y/R_0$')
plt.ylabel('$z/R_0$')
plt.title('Dehnen DF with PowerSphericalPotential, 1.7 Million Stars')
plt.show()
```

```python
bins = 200
```

```python
H, xedges, yedges = np.histogram2d(*centres.T[[0,1]], bins=bins, weights=max_dots) # Sum the max dot products
H /= np.histogram2d(*centres.T[[0,1]], bins=bins)[0] # Divide by number of points in each bin
H = H.T
X, Y = np.meshgrid(xedges, yedges)

plt.figure(figsize=(10,8))
plt.style.use("dark_background")
plt.pcolormesh(X, Y, H, cmap='plasma', vmin=0, vmax=1)
plt.colorbar(label='Average Maximum Absolute Dot Product')
plt.xlabel('$x/R_0$')
plt.ylabel('$y/R_0$')
plt.title('Dehnen DF with PowerSphericalPotential, 1.7 Million Stars')
plt.show()
```
