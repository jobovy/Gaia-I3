```python
import numpy as np
from scipy.interpolate import RectBivariateSpline as interpolation
```

```python
R_number = 10
z_number = 5
R_linspace = np.linspace(-5, 5, R_number)
z_linspace = np.linspace(-5, 5, z_number)
grid = np.empty((R_number, z_number))
for i, R in enumerate(R_linspace):
    for j, z in enumerate(z_linspace):
        grid[i][j] = R**2 + z**2
print(grid)
```

```python
ip = interpolation(R_linspace, z_linspace, grid)
```

```python
x = np.linspace(-5, 5, 1000)
y = np.linspace(-5, 5, 1000)
xv, yv = np.meshgrid(x, y)
z = ip.ev(xv, yv)
```

```python
%pylab inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z)
```
