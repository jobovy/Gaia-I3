```python
# McMillan2017 Rotation Curve
import warnings
warnings.filterwarnings('ignore')
import mcmillan
from galpy import potential
from galpy.util import bovy_plot

%pylab inline

mcm17= mcmillan.McMillan2017

# MWPotential2014 Rotation Curve
from galpy.potential import MWPotential2014

potential.plotRotcurve(MWPotential2014,Rrange=[0.01,10.],grid=1001,yrange=[0.,1.2],label="MWPotential2014")
potential.plotRotcurve(mcm17,Rrange=[0.01,10.],grid=1001,yrange=[0.,1.2],use_physical=False,label="McMillan2017",overplot=True)
title("Rotation Curve Comparison")
legend()
```

```python
#for mathew
print(potential.evaluatephiforces(MWPotential2014,R=0,z=0))
print(potential.evaluatephiforces(mcm17,R=0,z=0))
print(potential.evaluatephiforces(MWPotential2014,R=8,z=0))
print(potential.evaluatephiforces(mcm17,R=8,z=0))
print(potential.evaluatephiforces(MWPotential2014,R=8,z=0.1))
print(potential.evaluatephiforces(mcm17,R=8,z=0.1))
```
