```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('main_program_results/toomre_main/data.npz')
```

```python
cluster_centres = data['cluster']
dot_products = data['result']
```

```python
max_dot_products = np.nanmax(np.abs(dot_products), axis=1)
```

```python
cluster_centres = cluster_centres[~np.isnan(max_dot_products)]
max_dot_products = max_dot_products[~np.isnan(max_dot_products)]
```

```python
z = cluster_centres[:,2]
vz = cluster_centres[:,5]
```

```python
with plt.style.context(('dark_background')):
    plt.figure(figsize=(10,8), facecolor='black')
    plt.scatter(z, vz, c=max_dot_products, marker='.', s=5, cmap='plasma', vmin=0, vmax=0.5)
    plt.colorbar(label='Maximum Absolute Dot Product')
    plt.xlabel('$z$')
    plt.ylabel('$v_z$')
    plt.title('Maximum Absolute Value of Dot Product in $z-v_z$ Dimension')
    plt.show()
```
