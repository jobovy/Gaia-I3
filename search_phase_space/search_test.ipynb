$$\epsilon = 0.1$$

$$(x_0,y_0,z_0,U_0,V_0,W_0)=(0.1 \textrm{kpc}, 0.1\textrm{kpc}, 0, -7.5\textrm{km s}^{-1}, -20\textrm{km s}^{-1}, -7.5\textrm{km s}^{-1})$$

$$A=0.1$$

$$\sqrt{(x-x_o)^2 + (y-y_0)^2 + (z-z_0)^2 + (A(U - U_{0}))^2 + (A(V - V_{0}))^2 + (A(W - W_{0}))^2} < \epsilon$$

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(np.array(table['x']), np.array(table['y']), np.array(table['z']))
ax.set_xlabel('x (kpc)')
ax.set_ylabel('y (kpc)')
ax.set_zlabel('z (kpc)')
plt.show()
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(np.array(table['vx']), np.array(table['vy']), np.array(table['vz']))
ax.set_xlabel('U (km/s)')
ax.set_ylabel('V (km/s)')
ax.set_zlabel('W (km/s)')
plt.show()
```

```python
num_stars = []
v_scales = np.linspace(0.01,0.3,20)
for scale in v_scales:
    try:
        table = search_phase_space(0.1,0.1,0,-7.5,-20,-7.5,0.1,v_scale=scale)
        num_stars.append(len(table))
    except:
        break
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax.scatter(v_scales[:len(num_stars)], num_stars)
ax.set_xlabel('Velocity Scale Factor')
ax.set_ylabel('Stars Found')
plt.show()
```
